```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split

# Load the heart disease dataset
df = pd.read_csv('/content/heart_disease.csv')

# Display first 5 rows
df.head()
```

```python
# Check rows and columns
df.shape
```

```python
# Check column names and data types
df.info()
```

```python
# Check missing values in each column
df.isnull().sum()
```

```python
# Show statistical summary of numeric columns
df.describe()
```

```python
# Check target value counts
# 0 = No Heart Disease
# 1 = Heart Disease
df['target'].value_counts()
```

```python
# Fill missing values in numeric columns using mean value
df['trestbps'] = df['trestbps'].fillna(df['trestbps'].mean())
df['chol'] = df['chol'].fillna(df['chol'].mean())
df['thalach'] = df['thalach'].fillna(df['thalach'].mean())

# Check missing values again
df.isnull().sum()
```

```python
# Convert sex column into numeric values
# Male = 1, Female = 0

df['sex'] = df['sex'].str.lower()

label = LabelEncoder()
df['sex'] = label.fit_transform(df['sex'])

df.head()
```

```python
# Separate input features and target/output column

x = df.drop(columns='target', axis=1)
y = df['target']

print(x)
print(y)
```

```python
# Standardize input data

scaler = StandardScaler()
scaler.fit(x)

x_std = scaler.transform(x)

print(x_std)
```

```python
# Replace original input data with standardized data

x = x_std
```

```python
# Split dataset into training and testing data
# 80% data for training and 20% data for testing

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, stratify=y, random_state=2
)

print(x.shape, x_train.shape, x_test.shape)
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score

model = svm.SVC(kernel='linear')

model.fit(x_train, y_train)

# Training Accuracy
x_train_pred = model.predict(x_train)
train_acc = accuracy_score(y_train, x_train_pred)
print("Training Accuracy :", train_acc)

# Testing Accuracy
x_test_pred = model.predict(x_test)
test_acc = accuracy_score(y_test, x_test_pred)
print("Testing Accuracy :", test_acc)
```
